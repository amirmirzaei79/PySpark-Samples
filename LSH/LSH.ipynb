# Section A - Implimentation and Timing of LSH and Sequential Search
In this section sequential search (searching throw all the data) and lsh search (applying lsh and searching in candidate pairs (pairs with elements in same bucket)) are implemented and timed.

## Reducer function for search

```python
# Creating Reducer Function (to get closest n patches to each image)
def create_get_n_closest(n: int):
    # This function gets n as input and returns a function (reducer) that keeps n closest patches to each target
    def get_n_closest(closest, new):
        # This function keeps n closest patches to target
        # For each key (target) closest is the sorted (based on distance) list of each (patch_idx, distance)
        # For each new patch (new), this function finds it's place on list (if it should be in list) and adds it and moves rest of entries accordingly
        new = new[0]
        i = 0
        closest_len = len(closest)
        # This loops finds appropriate index for new
        while i < closest_len and closest[i][1] < new[1]:
            i += 1
        # If length of closest is still less than n, new could be just added in that index, 
        # else if index is less than n, other entries need to be shifted to make room
        if closest_len < n:
            closest.insert(i, new)
        elif i < closest_len:
            for j in range(n - 1, i, -1):
                closest[j] = closest[j - 1]
            closest[i] = new
        return closest

    return get_n_closest
```

## Exhaustive Search

```python
# Filtering Items with index 100, 200, 300, ..., 10000
# Creating Cartesian Product with other patches and filtering out rows with same patch twice (to create pairs)
# Calculating distances in each pair and saving result in form of (target_idx, [(patch_idx, distance)]) in map function (to be reduced later)
# Before adding differences for each elemnt to measure total distance, data is converted to int32 to account for possibility of larger distances 
# (distances are saved as int32 as well)
pairs = patches.filter(lambda x: x[0] % 100 == 0 and x[0] <= 1000).cartesian(patches).filter(lambda x: x[0][0] != x[1][0])
distances = pairs.map(
    lambda x: (x[0][0], [(x[1][0], np.absolute(x[0][1] - x[1][1]).astype(np.int32).sum())])
)
```

```python
# Use create_get_n_closest function to create appropriate reducer function and apply it on distances to get closest n (3) patches to each key (target)
search_results = distances.reduceByKey(create_get_n_closest(3))
```

```python
# flat map search_results (n closest patches to each target) to format (patch_idx, (target_idx, distance)) and join with patches to get data of each patch
# reformat entries (in map function) to ((target_idx, patch_idx), (patch_data, distance))
closest_images = search_results.flatMap(
    lambda row: [(x[0], (row[0], x[1])) for x in row[1]]
).join(patches).map(
    lambda row: ((row[1][0][0], row[0]), (row[1][1], row[1][0][1]))
)
```

## LSH
Includes Following Subsections:
* Setting LSH Constants
* Creation of Signiture Matrix
    * Creation of Random Hyperplanes
    * Signiture Matrix Creation
* LSH search

### LSH Constants

```python
r = 4
b = 5
bucket_length = 50
```

### Signiture Matrix
A random matrix is created and each column is normalized to length of 1
Signiture matrix is created by multiplying each vector (patch) by mentioned matrix (Projecting vector to a space with lower dimensionality)
And then dividing values of new vector by bucket length and converting result to uint16 (Putting values into buckets)

```python
random_vectors = np.random.random_sample((400, r * b)) # Results are from the “continuous uniform” over half open interval of [0.0, 1.0)
for i in range(r * b):
    random_vectors[:, i] /= np.sqrt(np.sum(random_vectors[:, i] ** 2)) # Normalizing length of vectors
```

```python
# create signiture matrix using map function and random matrix created
signiture_matrix = patches.mapValues(lambda x: (x.dot(random_vectors) / bucket_length).astype(np.uint16)).cache()
signiture_matrix.count()
```

### LSH

```python
# This function checks input pair to see if they are same in atleast one band
# Return (target_id, patch_id, Status) Status is True if pairs are similar, False if otherwise
def is_similar(x):
    for i in range(b):
        if np.array_equal(x[0][1][r*i:r*(i+1)], x[1][1][r*i:r*(i+1)]):
            return (x[0][0], x[1][0], True)
    return (x[0][0], x[1][0], False)
```

```python
# Filtering Items with index 100, 200, 300, ..., 10000
# Creating Cartesian Product with other signitures and filtering out rows with same signiture twice (to create pairs)
# Checking for each pair for lsh similarity and saving result in form of (target_id, patch_id, Status) in map function 
# Filtering pairs with true Status and reformat into (target_id, patch_id) (to create candidate pairs)

lsh_similar = signiture_matrix.filter(lambda x: x[0] % 100 == 0 and x[0] <= 1000).cartesian(signiture_matrix).filter(lambda x: x[0][0] != x[1][0]).map(
    is_similar
)

candidate_pairs = lsh_similar.filter(lambda x: x[2]).map(lambda x: (x[0], x[1]))
```

```python
# Joining candidate pairs with patches to get patches data for candidate pairs and saving into format ((target_id, target_data), (patch_id, patch_data))
candidate_patches = candidate_pairs.map(
    lambda x: (x[1], x[0])
).join(patches).map(
    lambda x: (x[1][0], (x[0], x[1][1]))
).join(patches).map(
    lambda x: ((x[0], x[1][1]), (x[1][0][0], x[1][0][1]))
)

# Calculating distances between candidate pairs
distances_lsh = candidate_patches.map(
    lambda x: (x[0][0], [(x[1][0], np.absolute(x[0][1] - x[1][1]).astype(np.int32).sum())])
)

# Use create_get_n_closest function to create appropriate reducer function and apply it on distances to get closest n (3) patches to each key (target)
search_results_lsh = distances_lsh.reduceByKey(create_get_n_closest(3))
```

```python
# flat map search_results_lsh (n closest patches to each target) to format (patch_idx, (target_idx, distance)) and join with patches to get data of each patch
# reformat entries (in map function) to ((target_idx, patch_idx), (patch_data, distance))
closest_images_lsh = search_results_lsh.flatMap(
    lambda row: [(x[0], (row[0], x[1])) for x in row[1]]
).join(patches).map(
    lambda row: ((row[1][0][0], row[0]), (row[1][1], row[1][0][1]))
)
```

# Section B - Comparing Results
In this section we find 10 closest matches to a target image using both methods and disply results and their measured distance from target

```python
# Getting target image & it's signiture
target_image = patches.filter(lambda x: x[0] == 100).cache()
target_image_signiture = signiture_matrix.filter(lambda x: x[0] == 100)
```

```python
# Putting All operations for sequential search (from previous section) together to find 10 closest patches to target
sequential_search_rdd = target_image.cartesian(patches).filter(lambda x: x[0][0] != x[1][0]).map(
    lambda x: (x[0][0], [(x[1][0], np.absolute(x[0][1] - x[1][1]).astype(np.int32).sum())])
).reduceByKey(create_get_n_closest(10)).flatMap(
    lambda row: [(x[0], (row[0], x[1])) for x in row[1]]
).join(patches).map(
    lambda row: ((row[1][0][0], row[0]), (row[1][1], row[1][0][1]))
)

sequential_search_result = sequential_search_rdd.collect()
```

```python
# Putting All operations for lsh search (from previous section) together to find 10 closest patches to target
lsh_search_rdd = target_image_signiture.cartesian(signiture_matrix).filter(lambda x: x[0][0] != x[1][0]).map(
    is_similar
).filter(lambda x: x[2]).map(
    lambda x: (x[0], x[1])
).map(
    lambda x: (x[1], x[0])
).join(patches).map(
    lambda x: (x[1][0], (x[0], x[1][1]))
).join(target_image).map(
    lambda x: ((x[0], x[1][1]), (x[1][0][0], x[1][0][1]))
).map(
    lambda x: (x[0][0], [(x[1][0], np.absolute(x[0][1] - x[1][1]).astype(np.int32).sum())])
).reduceByKey(create_get_n_closest(10)).flatMap(
    lambda row: [(x[0], (row[0], x[1])) for x in row[1]]
).join(patches).map(
    lambda row: ((row[1][0][0], row[0]), (row[1][1], row[1][0][1]))
)

lsh_search_result = lsh_search_rdd.collect()
```
